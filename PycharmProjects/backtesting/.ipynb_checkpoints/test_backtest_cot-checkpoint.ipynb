```python
import contract_utilities.expiration as exp
date_list = exp.get_bus_day_list(date_from=20100101, date_to=20180430)
#date_list.reverse()
date_list
```

```python
result_list = []
import opportunity_constructs.futures_medium as fm

for i in range(len(date_list)):
    print(date_list[i])
    output = fm.generate_fm_sheet_4date(date_to=date_list[i])
    output['futures']['settle_date'] = date_list[i]
    result_list.append(output['futures'])
```

```python
import pandas as pd
result_frame = pd.concat(result_list)
```

```python
# Univariate Analysis: Trend Direction

grouped = result_frame.groupby(['ticker_head','trend_direction'])

summary_frame = pd.DataFrame()
summary_frame['change1_dollar'] = grouped['change1_dollar'].mean()
summary_frame['change1_instant_dollar'] = grouped['change1_instant_dollar'].mean()
summary_frame['change5_dollar'] = grouped['change5_dollar'].mean()
summary_frame['change10_dollar'] = grouped['change10_dollar'].mean()
summary_frame['change20_dollar'] = grouped['change20_dollar'].mean()
summary_frame['trend_direction'] = grouped['trend_direction'].first()

positive_trend = summary_frame[summary_frame['trend_direction']==1]
positive_trend.mean()

# Trend Direction seems to have a small mean reverting effect
```

```python
# Univariate Analysis: RSI

result_frame['oversold_overbought'] = 0

result_frame['oversold_overbought'].loc[result_frame['rsi_14']>80] = 1
result_frame['oversold_overbought'].loc[result_frame['rsi_14']<20] = -1

grouped = result_frame.groupby(['ticker_head','oversold_overbought'])

summary_frame = pd.DataFrame()
summary_frame['change1_dollar'] = grouped['change1_dollar'].mean()
summary_frame['change1_instant_dollar'] = grouped['change1_instant_dollar'].mean()
summary_frame['change5_dollar'] = grouped['change5_dollar'].mean()
summary_frame['change10_dollar'] = grouped['change10_dollar'].mean()
summary_frame['change20_dollar'] = grouped['change20_dollar'].mean()
summary_frame['oversold_overbought'] = grouped['oversold_overbought'].first()

positive_trend = summary_frame[summary_frame['oversold_overbought']==1]
positive_trend.mean()

# Some evidence of mean reversion although it's not very consistent
```

```python
# Univariate Analysis: Curve Slope

result_frame['curve_slope_direction'] = 0

result_frame['curve_slope_direction'].loc[result_frame['curve_slope']>70] = 1
result_frame['curve_slope_direction'].loc[result_frame['curve_slope']<30] = -1

grouped = result_frame.groupby(['ticker_head','curve_slope_direction'])

summary_frame = pd.DataFrame()
summary_frame['change1_dollar'] = grouped['change1_dollar'].mean()
summary_frame['change1_instant_dollar'] = grouped['change1_instant_dollar'].mean()
summary_frame['change5_dollar'] = grouped['change5_dollar'].mean()
summary_frame['change10_dollar'] = grouped['change10_dollar'].mean()
summary_frame['change20_dollar'] = grouped['change20_dollar'].mean()
summary_frame['curve_slope_direction'] = grouped['curve_slope_direction'].first()

positive_trend = summary_frame[summary_frame['curve_slope_direction']==1]
positive_trend.mean()

# There's some consistency with this
```

```python
# Univariate Analysis: cot_index

result_frame['cot_direction'] = 0

result_frame['cot_direction'].loc[result_frame['comm_cot_index_fast']>80] = 1
result_frame['cot_direction'].loc[result_frame['comm_cot_index_fast']<20] = -1

grouped = result_frame.groupby(['ticker_head','cot_direction'])

summary_frame = pd.DataFrame()
summary_frame['change1_dollar'] = grouped['change1_dollar'].mean()
summary_frame['change1_instant_dollar'] = grouped['change1_instant_dollar'].mean()
summary_frame['change5_dollar'] = grouped['change5_dollar'].mean()
summary_frame['change10_dollar'] = grouped['change10_dollar'].mean()
summary_frame['change20_dollar'] = grouped['change20_dollar'].mean()
summary_frame['cot_direction'] = grouped['cot_direction'].first()

positive_trend = summary_frame[summary_frame['cot_direction']==1]
positive_trend.mean()

# Not very consistent results
```

```python
cot_output = output
comm_net = cot_output['comm_net'].iloc[-1]
comm_net_min_slow = cot_output['comm_net'].iloc[-156:].min()
comm_net_max_slow = cot_output['comm_net'].iloc[-156:].max()

comm_cot_index_slow = 100*(comm_net-comm_net_min_slow)/(comm_net_max_slow-comm_net_min_slow)

comm_net_min_fast = cot_output['comm_net'].iloc[-52:].min()
comm_net_max_fast = cot_output['comm_net'].iloc[-52:].max()

comm_cot_index_fast = 100*(comm_net-comm_net_min_fast)/(comm_net_max_fast-comm_net_min_fast)

print(comm_cot_index_slow)
print(comm_cot_index_fast)
```
