```python
import matplotlib.pyplot as plt

plt.plot(range(len(data_out.index)),data_out['Node 2']
        ,range(len(data_out.index)),data_out['Node 1'])
plt.grid()         
         
plt.show()
```

```python
import shared.statistics as stats
from sklearn.decomposition import PCA

n_components = 2

pca = PCA(n_components=n_components)

data_input = data_out[['Node 1','Node 2','Node 3','Node 4','Node 5','Node 6']]

pca_out = stats.get_pca(data_input=data_input, n_components=n_components)

scores = pca.fit_transform(data_input)
loadings = pca.components_
inverse_data = pca.inverse_transform(scores)
explained_variance = pca.explained_variance_ratio_
print('Explained Variance Ratio: ' + str(sum(explained_variance)))
```
